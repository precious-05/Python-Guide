```python
# Example DataFrame with different non-null counts
import pandas as pd
import numpy as np

data = {
    'age': [25, 30, np.nan, 22],
    'gender': ['F', 'M', 'F', np.nan],
    'occupation': ['Engineer', 'Doctor', 'Sing', 'Artist'],
    'zip_code': ['12345', '67890', '54321', '98765']
}

df = pd.DataFrame(data)
print(df)
print("\n")
# Get non-null counts for each column
counts = df.count()

# Maximum count value
#df.count().max() gives the maximum count of non-null values across columns.
max_count = counts.max()

# Column name with maximum non-null values
#df.count().idxmax() returns the name of the first column with this maximum count.
column_with_max_count = counts.idxmax()

print(f"Maximum number of non-null values: {max_count}")
print(f"Column with first occurrence of max non-null values: {column_with_max_count}")





                      # non-null counts for each column == maximum count of non-null values across columns
col_with_max_count = counts[counts == max_count].index.tolist()
print(f"Columns with max non-null values: {col_with_max_count}")


#Boolean Indexing: counts[counts == max_count] filters the Series based on a condition.
#.index: Extracts the index of the filtered Series.
#.tolist(): Converts the index into a list
```

how idxmax() works in conjunction with value_counts() and how they relate to sorting:

value_counts() and idxmax()
1. value_counts()
Purpose: It counts the occurrences of each unique value in a Series and returns a Series where:
The index is the unique values from the original Series.
The values are the counts of these unique values.
Sorting: By default, value_counts() sorts the results in descending order, with the most frequent value appearing first.


So, when using value_counts(), the Series is sorted by the counts, and idxmax() will return the label of the most frequent value (or the first one if there is a tie)

such as there is Engineer and Doctor having same counts 2 but top most is engineer so idmax returns it only

```python
import pandas as pd

# Create a DataFrame
users = pd.DataFrame({
    'occupation': ['Engineer', 'Doctor', 'Engineer', 'Artist', 'Doctor']
})


#print(occupation_counts)

# Find the most frequent occupation
most_frequent_occupation = users['occupation'].value_counts().idxmax()
print(f"The most frequent occupation is: {most_frequent_occupation}")

#using value counts will gibe the frequency of each value
# but using idxmax() will give the lable of the value having highest frequency

least_frequent_occupation= users['occupation'].value_counts().idxmin()
print(f"\nThe least frequent occupation is: {least_frequent_occupation}")


#users['occupation'].value_counts()
```
