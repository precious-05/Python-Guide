- count: The number of non-missing (non-null) values in each column. For example, count for "Age" is 383, indicating that there are 383 non-null values in the "Age" column.

- unique: The number of unique values in each column. For categorical columns (like "Gender" and "Smoking"), this shows how many distinct categories are present.





- top: The most frequently occurring value in the column. For categorical columns, this shows the most common category. For example, the "Gender" column has 'F' as the most common value.

- freq: The frequency of the most common value (i.e., how many times the value in the "top" row appears). For example, 'F' appears 312 times in the "Gender" column.

- mean: The average value of the column, applicable to numerical columns. For example, the mean age is approximately 40.87. This is not applicable to categorical columns, so those will show NaN (Not a Number).

- std: The standard deviation, which measures the spread or dispersion of the numerical values in the column. For the "Age" column, the standard deviation is approximately 15.13. This is also not applicable to categorical columns.

- min: The minimum value in the column for numerical data. For the "Age" column, the minimum age is 15.

- 25%: The 25th percentile value, meaning that 25% of the data is less than or equal to this value. For the "Age" column, 25% of the ages are 29 or younger.

- 50%: The median value, meaning that 50% of the data is less than or equal to this value. For "Age," the median age is 37.

- 75%: The 75th percentile value, meaning that 75% of the data is less than or equal to this value. For "Age," 75% of the ages are 51 or younger.

- max: The maximum value in the column for numerical data. For the "Age" column, the maximum age is 82.

- For Categorical Columns:
- unique, top, and freq are relevant, showing the diversity and most common values.
- mean, std, min, 25%, 50%, 75%, and max are not applicable and will show NaN.

```python
import pandas as pd # type: ignore
import numpy as np
dataset='Thyroid_Diff.csv'
patient=pd.read_csv(dataset)
numeric_cols = patient.describe()
print(numeric_cols)
mean_dict = {}
mode_dict = {}
median_dict = {}
std_dev_dict = {}
variance_dict = {}
max_dict = {}
min_dict = {}


for col in numeric_cols:
    mean_dict[col] = patient[col].mean()
    #mean_dict['Age']=patient['Age'].mean()
    mode_dict[col] = patient[col].mode().values[0]
    #mode_dict['Age'] = patient['Age'].mode().values[0]
    #values[0] is used to extract the first value from the Series returned by .mode() 
    # to handle cases where multiple modes might be returned, ensuring you get a single value for simplicity.
    median_dict[col] = patient[col].median()
    #median_dict['Age'] = patient['Age'].median()
    std_dev_dict[col] = patient[col].std()
    #std_dev_dict['Age'] = patient['Age'].std()
    variance_dict[col] = patient[col].var()
    #variance_dict['Age'] = patient['Age'].var()
    max_dict[col] = patient[col].max()
    #max_dict['Age'] = patient['Age'].max()
    min_dict[col] = patient[col].min()
    #min_dict['Age'] = patient['Age'].min()
    


print("Mean:\n", mean_dict)
print("\nMode:\n", mode_dict)
print("\nMedian:\n", median_dict)
print("\nStandard Deviation:\n", std_dev_dict)
print("\nVariance:\n", variance_dict)
print("\nMaximum:\n", max_dict)
print("\nMinimum:\n", min_dict)    
 
 
    
```
