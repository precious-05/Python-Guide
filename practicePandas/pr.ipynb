```python
import pandas as pd

df = pd.read_csv('all_data.csv')

# Drop rows with NaN values
df.dropna(inplace=True)

# Reset the index
df.reset_index(drop=True, inplace=True)

# Split 'Order Date' into 'Date' and 'Time'
df[['Date', 'Time']] = df['Order Date'].str.split(' ', n=1, expand=True)

# Split the 'Date' column into 'month', 'Days', and 'year'
df[['month', 'Days', 'year']] = df['Date'].str.split('/', n=2, expand=True)

# Delete the 'Date' column
del df['Date']

# Convert 'month' column to numeric, coercing errors to NaN
df['month'] = pd.to_numeric(df['month'], errors='coerce')

# Group by 'month' and sum the 'Quantity Ordered' column, showing the top 5
#print(df.groupby('month')['Quantity Ordered'].sum().nlargest(5))

# Alternatively, sort all month sums in descending order
print(df.groupby('month')['Quantity Ordered'].sum().sort_values(ascending=False))
```
