```python
import pandas as pd
Data=pd.read_excel('Customer-Call.xlsx')
Data
```

```python
Data.shape
```

```python
#Data=Data.dropna(subset=["Additional Assignment"])
Data=Data.dropna(how="all",axis=1) 
#how = any means agr row mn koi ek bhi value missing hai tb row drop hgi
#How=all ka mtlb hai agr row mn sb values NaN hngi tb row drop hogi
Data
```

```python
Data.isnull().sum()
```

```python
Data=Data.dropna(how="any",axis=1) #by default any
Data
```
