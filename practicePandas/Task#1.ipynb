```python
#Import the necessary libraries
import pandas as pd
#Import the dataset 

link='https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'

#Assign it to a variable
data=pd.read_csv(link,sep='\t')

#How many products cost more than $10.00?
#Since the item_price column is of object type , String or text...
# we can't directly perform numerical comparisons on String such as here checking if a price is greater than 10.00...
# Replacing the $ sign using replace function , and converting the dtype of "item_price" column into float using astype() method
#data.dtypes
print(pd.__version__)
```

#### When You Need regex=True:
- Complex Patterns: If you need to find and replace text based on a pattern rather than a specific string, you'll use regex=True.
**Examples:**
- Replacing all sequences of digits (e.g., r'\d+').
- Replacing all words that start with a specific letter.
- Replacing based on conditions like the end of a string ($) or beginning of a string (^).
#### When You Don't Need regex=True:
- Simple Replacements: If you're just replacing specific characters or sequences of characters without any special pattern, you usually don’t need to worry about regex.
**Examples:**
- Removing a dollar sign ('$').
- Replacing a specific word or phrase with another.

```python
#data.dtypes
#data['item_price'] = data['item_price'].replace('$', '', ).astype(float)  #wrong without str accessor
#data['item_price'] = data['item_price'].str.replace('[$]', '', regex=True).astype(float)
data['item_price'] = data['item_price'].str.replace('$', '').astype(float)

#data.dtypes

#data['item_price'].dtypes
#data['item_price']
```

```python
cost=data[data['item_price']>10.00]
cost
```

```python
#How many products cost more than 10$
len(cost['item_price'].value_counts())
```
