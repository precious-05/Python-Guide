- _Pandas Visualization: Easy to use interface, built-in on Matplotlib_

```python
import pandas as pd
from matplotlib import pyplot as plt  #pyplot is a class
# In jupyter Python Interpreter is Known as Kernel
# Bar Plot
df=pd.read_csv('C:\\Users\\Hp\\Desktop\\learning\\Python-Guide\\practicePandas\\thyroid2.csv')
```

#### **Bar Plot**

```python
x=df['Age']
y=df['Recurred']
plt.figure(figsize=(10,4))
plt.xlabel('Age',fontsize=20)
plt.ylabel('Recurred',fontsize=20)
plt.title('Age Vs Recurred Bar Plot',fontsize=20)
#  Giving each bar a different Color
clr=['#7bdff2','#f7d6e0']
plt.bar(x,y,width=0.7,color=clr,align='center',edgecolor='Black',linewidth=0.8,alpha=1)  #--- giving each bar a single color only as '#f057'
# To give each bar a different color we pass a list of colors not a ingle color
#The value of alpha is b/w 0 and 1 only
#align is used to align the text below bars such as in my case these are numbers such as 1.0 , 1.5 to center or at edge
#edgecolor is used for the border color of bars
#linewidth / linestyle are used for the the border with or style of borders of bars
#plt.bar(x,y,width=0.3,color=clr,align='edge')  #--- giving each bar a single color only as '#f057'
#plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,4))
x=df['Age']
y=df['Risk']
plt.xlabel('Age',fontsize=20)
plt.ylabel('Risk',fontsize=20)
plt.title('Age vs Risk',fontsize=20)
clr2=['#50808e','#84b59f']
plt.bar(x,y,color=clr2,align='edge',width=0.7,edgecolor='Black',alpha=1,linewidth=0.7)  #,label='Ages'
#plt.legend()
plt.show()

#  The bars represent the risk level associated with a certain age.
# The color of each bar represents a different category:
#Darker Blue: Represents the risk for an individual in that age group.
#Lighter Green: Represents the average risk for all individuals in that age group.
#So, each bar shows the risk for that particular age, with the darker blue bar showing the individual's risk and 
# the lighter green bar showing the average risk for that age group.
```
