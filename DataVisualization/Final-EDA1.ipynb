### **Univariate Analysis**

```python
# uni means one & variate means variable & variable means column
# In univariate analysis we are taking 1 variable means 1 column at a time 
# In this we perform analysis without considering relationships with other variables

category_counts = df['gender'].value_counts()

plt.figure(figsize=(8, 4))
category_counts.plot(kind='bar', color='#9d4edd',edgecolor='Black')
plt.title(f'Univariate Analysis of gender Column',fontsize=15)
plt.xlabel('Gender',fontsize=15)
plt.ylabel('Frequency',fontsize=15)
plt.xticks(rotation=45)
plt.grid(axis='y',linestyle='--',alpha=0.6)

plt.show()

#This univariate analysis shows the frequency of each gender in a dataset. 
# It appears that there are more females (F) than males (M) in the dataset. 
# The frequency of females is almost three times the frequency of males.
```

```python
# Using pd.cut() to create bins
bins = pd.cut(df['age'], bins=[15, 30, 45, 60, 75, 82], labels=['15-30', '31-45', '46-60', '61-75', '76-82'])

# Plotting the value counts of each bin
#The bar plot will show how many ages fall into each bin.

plt.figure(figsize=(8, 4))
bins.value_counts().sort_index().plot(kind='bar', color=['#e09f3e', '#2ec4b6'], edgecolor='black')

# Customizing the plot
plt.title('Age Distribution with Bins')
plt.xlabel('Age Group')
plt.ylabel('Frequency')
plt.xticks(rotation=45)  # Rotating x-axis labels for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Adding grid for clarity

plt.show()




```

#### **Age Distribution Analysis: Key Insights**
- The bar plot above represents the age distribution within the dataset, divided into 5 groups: 15-30, 31-45, 46-60, 61-75, and 76-82. The most frequent group is 31-45, with approximately 140 individuals, while the 76-82 age group has the least representation, with fewer than 20 individuals. There is a clear trend showing a decline in frequency as the age increases. Most of the dataset comprises individuals aged between 15 and 60, indicating that the population of patients skews towards younger and middle-aged individuals
