##### ```Why we need Data Visualization and what is Data Visualization```
- We need data visualization because human brain can process information easily when it is in pictorial or graph form. 

- Example: If a person is running his busines in Pakistan in different Cities. Suppose his business has good sales in Islamabad and Lahore. Now if he finds this sales using numbers then it will be difficult for him to understand because there is a lot of data, a lot of customers data.

- So it becomes difficult to detect that in which city there is the top most sale using numbers. But it will be easy to detect using the chart. such as making a bar graph. In the bar graph if the bar of Islamabad is larger than the bar of Karachi then it will be very easy to understand that the sales in Islamabad is good. Human brain will easily capture this because of colors, size e.t.c

- That's why visualization is very important

- Why a customer will choose or prefer the service of a specific businessman.The business is dynamic. There are many Competitors availaible.To buy 1 product there are 10 shops/business are available.Then why a customer will choose a specific brand / business.It is because he is providing a unique service by understanding his data like what to sale how to sale , that this customer wants this product more so he has to sale him this

- Data Visualization allows us to quickly interpret the data and adjust different variables to see their effect

- Data Visualization is the presentation of the data in pictorial or graphical format

- Processes: > Transform dataset , Document insight , Analyze , Visualize.

``` What is Matplotlib ?? ```
- It is a Python package used for 2D graphs such as making bar chart, line chart, pie chart, scatter plot e.t.c

1: _Bar Graph: If there is any Category such as gender sales (male/female) or region sales or profit or salary we use bar plot_

2: _Histogram: It tells about data distribution such as how many customers are b/w age of 0-5 years_

3: _Scatter plot: If we want to compare 2 things such as profit or sales_

4: _Pie-Chart: It shows the composition such as in which region there is how much sale example: in north there is 2000$ sale in west its 1000$ sale_

5: _Hexagonal Bin Plot: It shows teh density such as if there are more customers with age b/w 20-30 then the density will show them_

6: _Area plot: It is plot against time. That how sales are going with time_
